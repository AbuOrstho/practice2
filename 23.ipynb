Задание 1. Используйте библиотеку Numpy

Создайтё массив нулей длины 30, массив единиц длины 30 и массив чисел от -15 до 14.

Сложите три массива, напечатайте ответ.

Прибавьте к массиву 3, напечатайте ответ.


```python
import numpy as np


first_array = np.zeros(30)
second_array = np.ones(30)
third_array = np.arange(-15, 15)

print(first_array, second_array, third_array, sep='\n')

answer = first_array + second_array + third_array
print(answer)
print(answer + 3)
```

Задание 5. Используйте matplotlib.pyplot

Создайте массив чисел от 0 до 100 и массив чисел от 50 до -50.

Напечатайте график, на котором эти числа будут изображены в виде точек. Не забудьте задать нужные абсциссы.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 101)
y = np.arange(50, -51, -1)

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('я так мучался с этим юпитером')
plt.show()
```

Задание 9. Используйте matplotlib.pyplot

Напечатайте график функции sin x на отрезке от -3.14 до 3.14. 


```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-3.14, 3.14)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

Задание 9. Используйте matplotlib.pyplot

Напечатайте график функции cos x на отрезке от -3.14 до 3.14. Добавьте графику сетку, заголовок и легенду с названием функции.


```python
plt.figure()
x = np.linspace(-3.14, 3.14, 100)
plt.plot(x, np.cos(x))
plt.legend('f')
plt.grid()
plt.title('cos x')
plt.show()
```

Задание 10*. Используйте matplotlib.pyplot

Напечатайте круг с центром в (0,0) и радиусом 1.

```python
fig, ax = plt.subplots()
circle = plt.Circle((0, 0), 1, color='b')
circle1 = plt.Circle((0, 0), 0.5, color='r')
ax.add_patch(circle)
ax.add_patch(circle1)
plt.axis('scaled')
plt.show()
```
